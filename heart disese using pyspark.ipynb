## read data with pyspark

###data processing

```python
x=['age','sex','trestbps','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']
y=['num']
full_coll=['age','sex','trestbps','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal','num']
```

```python
from pyspark.ml.feature import VectorAssembler
va=VectorAssembler(inputCols=x,outputCol='data_merge')
df=va.transform(df)
df.show(3)
```

```python
dataset=df.select(['data_merge','num'])
dataset.show(5)
```

```python
train_set,test_set=dataset.randomSplit([0.75,0.25])
```

```python
from pyspark.ml.classification import RandomForestClassifier, NaiveBayes, DecisionTreeClassifier
```

```python
RF=RandomForestClassifier(featuresCol='data_merge',labelCol='num',numTrees=50)
NI = NaiveBayes(featuresCol='data_merge', labelCol='num', modelType='gaussian')
DT=DecisionTreeClassifier(featuresCol='data_merge',labelCol='num')
```

```python
RF=RF.fit(train_set)
NI=NI.fit(train_set)
DT=DT.fit(train_set)
```

```python
rf_pred=RF.transform(test_set)
ni_pred=NI.transform(test_set)
dt_pred=DT.transform(test_set)
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator
eval = BinaryClassificationEvaluator(labelCol='num')
```

```python
accu_random=eval.evaluate(rf_pred)
accu_score_naive=eval.evaluate(ni_pred)
accu_score_desition=eval.evaluate(dt_pred)
```
