```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## ADD data

```python
df=pd.read_csv('G:/ترم 7/heart_disease_dataset.csv' )
df
```

## because data have label we are drop num column to change data to unsuprevised learning

```python
df.drop('num', axis=1, inplace=True)
df
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=2)
km.fit(df)
labels = km.predict(df)
labels
```

```python
km.inertia_
```

```python
xs = df.chol
ys = df.thalach
centroids = km.cluster_centers_
plt.scatter(xs, ys, c=labels)
plt.scatter(centroids[:,4],centroids[:,7],marker='x',s=150,alpha=0.5)
plt.show()
```

```python
print(km.inertia_)
```

```python
from scipy.cluster.hierarchy import linkage,dendrogram,fcluster
import matplotlib.pyplot as plt
hir = linkage(df,method='complete')
labels = fcluster(hir, 200, criterion='distance')
print(labels)
dendrogram(hir,leaf_rotation=90)
plt.show()
```

```python
hir = linkage(df,method='complete')
labels = fcluster(hir, 120000, criterion='distance')
print(labels)
dendrogram(hir,leaf_rotation=60)
plt.show()
```

```python
from sklearn.cluster import MeanShift
x = df
ms = MeanShift(bandwidth=10)
ms.fit(x)
labels = ms.labels_
cluster_center = ms.cluster_centers_
n_cluster = len(np.unique(labels))
print('Number of estimated cluster:' ,n_cluster)
plt.scatter(df.chol,df.thalach, c=labels)
plt.scatter(cluster_center[:,4], cluster_center[:,7], marker='x', s=150, linewidth=3)
plt.show()
```
